##This notebook contains some of the visualization styles using seaborn package.

```python
#import required packages:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#assign the file:
file = 'Downloads/auto-mpg.csv'

#import the file into dataframe:
df = pd.read_csv(file)

#use subplot1():
plt.subplot(2, 1, 1)
sns.stripplot(x='cyl', y='hp', data=df)

#use subplot2():
plt.subplot(2, 1, 2)
sns.stripplot(x='cyl', y='hp', data=df, size=5, jitter=True)

#show the plot:
plt.show()
```

##2D array pseudo color

```python
#assign the file:
file = 'Downloads/coupe.jpg'

#read the file:
img = plt.imread(file)

#shape of the image:
print(img.shape)

intensity = img.sum(axis=2)

print(intensity.shape)

plt.imshow(intensity, cmap='gray')

plt.axis('off')

plt.show()
```

##simple linear regressions

```python
sns.lmplot(x='weight', y='hp', data=df)

plt.show()
```

##residual of regressions

```python
#use residplot:
sns.residplot(x='hp',y='mpg', data=df, color='green')

plt.show()
```

##high order regressions:

```python
#scatter plot for weight & mpg:
plt.scatter(df['weight'], df['mpg'], label='data', color='red', marker='o')

#reg plot for order 1
sns.regplot(x='weight', y='mpg', data=df, scatter=None, color='blue', label='order1')

#reg plot for order 2
sns.regplot(x='weight', y='mpg', data=df, scatter=None, color='green', label='order2')

#axis:
plt.legend(loc='upper right')

plt.show()
```

##visualize univariate distributions

```python
#swarmplots: a
plt.subplot(2,1,1)

sns.swarmplot(x='cyl', y='hp', data=df)

plt.subplot(2,1,2)

sns.swarmplot(x='cyl', y='hp', data=df, hue='origin', orient='v')

plt.show()
```

##Violin plots

```python
#subplot 1:
plt.subplot(2,1,1)

#violinplot 1:
sns.violinplot(x='cyl', y='hp', data=df)

#subplot 2:
plt.subplot(2,1,2)

#violin plot2:
sns.violinplot(x='cyl', y='hp', data=df, inner=None, color='gray')

#overlaying a strip plot on violin plot:
sns.stripplot(x='cyl', y='hp', data=df, jitter=True, size=1.5)

plt.show()
```

##Histogram equilization in images

```python
#load the image in an array:
file = 'Downloads/coupe.jpg'

#assign file:
img = plt.imread(file)

#subplot 1:
plt.subplot(2,1,1)

plt.title('Original image')

plt.axis('off')

#print the image:
plt.imshow(img, cmap='gray')

#flatten the image:
pixels = img.flatten()

#display the histogram:
plt.subplot(2,1,2)

plt.xlim((0,255))

plt.title('normalized histogram')

plt.hist(pixels, bins=64, color='red', alpha=0.4, range=(0,256), normed=True)

plt.show()
```

##extract histograms from color image:

```python
plt.subplot(2,1,1)

plt.title('Original image')

plt.axis('off')

plt.imshow(img)

red, green, blue = img[:,:,0], img[:,:,1], img[:,:,2]

#flatten the 2D arrays for RGB channels:
red_pixels = red.flatten()
green_pixels = green.flatten()
blue_pixels = blue.flatten()

#overlay of histogram of pixels of each other in bottom subplots:
plt.subplot(2,1,2)

plt.title('Histogram of color images')

plt.xlim((0,256))

#red:
plt.hist(red_pixels, bins=64, normed=True, color='red', alpha=0.2)

#green
plt.hist(green_pixels, bins=64, normed=True, color='green', alpha=0.2)

#blue:
plt.hist(blue_pixels, bins=64, normed=True, color='blue', alpha=0.2)

plt.show()
```
